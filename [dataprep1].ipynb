```python

from sklearn.metrics import mean_absolute_percentage_error as MAPE
import numpy as np

df2 = pd.read_excel('คำตอบนักประเมิน.xlsx',sheet_name= 'ที่ดิน',skiprows=0)
ld1 = df2[df2['ID ผู้ประเมิน']==1]
ld1 = ld1.dropna(subset=['ราคาทรัพย์','ประเมินราคาซื้อ-ขาย'])
ld2 = df2[df2['ID ผู้ประเมิน']==2]
ld2 = ld2.dropna(subset=['ราคาทรัพย์','ประเมินราคาซื้อ-ขาย'])
ld2.rename(columns = {'ประเมินราคาซื้อ-ขาย' : 'ประเมิน2'}, inplace = True)
ld2['ประเมิน1'] = ld2['post id'].map(ld1.set_index('post id')['ประเมินราคาซื้อ-ขาย'])
all2 = ld2[['post id','ราคาทรัพย์','ประเมิน1','ประเมิน2']]
all2['check1']=all2.apply(lambda x:check_v(x['ราคาทรัพย์'],x['ประเมิน1']),axis=1)
all2['check2']=all2.apply(lambda x:check_v(x['ราคาทรัพย์'],x['ประเมิน2']),axis=1)

all2['ประเมิน1'] = np.where(all2['check1'] > 10, all2['ประเมิน1']*10, all2['ประเมิน1'])
all2['ประเมิน2'] = np.where(all2['check2'] > 10, all2['ประเมิน2']*10, all2['ประเมิน2'])

ld1MAPE = MAPE(all2['ประเมิน1'],pd.to_numeric(all2['ราคาทรัพย์']))
ld2MAPE = MAPE(all2['ประเมิน2'],pd.to_numeric(all2['ราคาทรัพย์']))
print(f'ความแตกต่างระหว่างราคาประกาศขายและราคาประเมินของ\nที่ดิน\nนักประเมิน คนที่ 1 = {ld1MAPE*100:.2f}%\nนักประเมิน คนที่ 2 = {ld2MAPE*100:.2f}%')
all2 = ld2.dropna(subset=['ประเมิน2','ประเมิน1','ราคาทรัพย์'])
one2two = MAPE(all2['ประเมิน1'],all2['ประเมิน2'])
two2one = MAPE(all2['ประเมิน2'],all2['ประเมิน1'])
```
